##Lambda Functions

- Anonymous Function
- Unnamed Small Functions

```python
square=lambda n : n**2
print(square(4))
```

```python
number=lambda n : "even" if n%2==0 else "odd"     #This is called ternary operator
print(number(4))
```

```python
cube=lambda n : n**3
print(cube(3))
```

```python
capitalize=lambda word : word.capitalize()
print(capitalize("dheeraj"))
```

```python
maximum_value=lambda a,b,c: max(a,b,c)
print(maximum_value(1,2,3))
```
