## Lambda Functions

- Anonymous Function
- Unnamed Small Functions

```python
square=lambda n : n**2
print(square(4))
```

```python
number=lambda n : "even" if n%2==0 else "odd"     #This is called ternary operator
print(number(4))
```

```python
cube=lambda n : n**3
print(cube(3))
```

```python
capitalize=lambda word : word.capitalize()
print(capitalize("dheeraj"))
```

```python
maximum_value=lambda a,b,c: max(a,b,c)
print(maximum_value(1,2,3))
```

## File Handling

- Readiing file in Python
- Writing in file in Python

```python
import csv

# for writing in the file

people=[
    ['Dan',34,'Bucherest'],
    ['Andrei',21,'London'],
    ['Maria',45,'Paris']
]

with open("people1.csv","w", newline="") as file:
    writer= csv.writer(file)
    writer.writerows(people)

# For reading the file

with open("people1.csv","r") as file:
    reader_object=csv.reader(file)
    print(list(reader_object))

    # next(reader_object)    If we want to directly jump to second element from the list
    for row in reader_object:
        print(row)
```
