## ChatBot


````python
from google import genai
from google.colab import userdata

client = genai.Client(api_key=userdata.get('GEMINI_API_KEY'))

History=[]
while True:

  prompt=input("Enter the prompt: ")
  if prompt.strip().lower()=="exit":
    print("Good Bye!")
    break
  else:
    History.append(f"User_input:{prompt}")
    response = client.models.generate_content(
      model="gemini-3-flash-preview",
      contents=History)
    print(response.text)
````

## Topic-->Detailed Content-->5 point Summary

````python
from google import genai
from google.colab import userdata

client = genai.Client(api_key=userdata.get('GEMINI_API_KEY'))

Topic=input("Enter the Topic: ")

prompt=f"Give me the detailed content on the topic:{Topic}"

response = client.models.generate_content(
    model="gemini-3-flash-preview",
    contents=prompt,
)
Data=response.text
print(Data)
````

````python
Summary=f"Give me the 5 point summary on the given Data:{Data}"

response = client.models.generate_content(
    model="gemini-3-flash-preview",
    contents=Summary,
)
print(response.text)
````
