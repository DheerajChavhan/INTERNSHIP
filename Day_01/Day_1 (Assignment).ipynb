✅ Problem 1: Collect Student Details (Variables + Input + List)
Write a Python program that:

Takes the student name as input

Takes marks for 3 subjects as input

Stores the marks inside a list called marks

Prints the name and marks list

```python
Student=input("Enter Student Name: ")

S1=int(input("Enter marks of Subject_1: "))
S2=int(input("Enter marks of Subject_2: "))
S3=int(input("Enter marks of Subject_3: "))

Marks=[S1,S2,S3]
print(f"Marks of {Student} are {Marks} ")
```

✅ Problem 2: Calculate Total and Average (Loop + List)

Using the marks list from Problem 1:

Use a loop to calculate the total marks

Calculate the average marks

Print both total and average

```python
for i in Marks:
  Total=sum(Marks)
  Average=Total/len(Marks)

print(f"Total Marks:{Total}")
print(f"Average Marks:{Average}")
```

✅ Problem 3: Decide Pass or Fail (Conditionals + Loop)

Using the marks list:

Check if the student has scored 35 or above in all subjects

If yes → print "Pass"

If no → print "Fail"

```python
Pass_Status=True
for i in Marks:
  if i <35:
    Pass_Status=False

if Pass_Status==True:
  print("Pass")
else:
  print ("Fail")
```

✅ Problem 4: Create a Student Dictionary (Dictionary + Key-Value Storage)

Create a dictionary called student that stores:

"name" → student name

"marks" → list of marks

"total" → total marks

"average" → average marks

"result" → pass/fail

Print the full dictionary.

```python
Student={
    "name":Student,
    "Marks":Marks,
    "Total":Total,
    "Average":Average,
    "Result":Pass_Status
}
print(Student)

```
