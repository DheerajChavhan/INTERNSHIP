```python
text=input("Enter The Text: ")
def get_word_count(text):
    word=text.split()
    return len(word)

get_word_count(text)
```

```python
text=input("Enter The Text: ")
def get_char_count(text):
  no_space_char = text.replace(" ","")
  char_count=len(no_space_char)
  return char_count

get_char_count(text)
```

```python
from langchain_core.runnables import RunnableLambda

#Converting Simple python function to a runnable function
word_count=RunnableLambda(get_word_count)
char_count=RunnableLambda(get_char_count)
```

```python
from langchain_core.runnables import RunnableParallel

Parellel=RunnableParallel({
    "Words":word_count,
    "Characters":char_count
})
result=Parellel.invoke(text)
print(result)
```

```python
from tavily import TavilyClient
from google.colab import userdata

def web_search(query):
  client = TavilyClient(userdata.get('TAVILY'))

  response = client.search(
      query=query,
      include_answer="basic",
  )
  return(response.get("answer"))
print(web_search("Who will win the ipl 2026"))
```
