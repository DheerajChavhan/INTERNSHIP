```python
# Import ChatModel
from google import genai
from google.colab import userdata
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser

City=input("Enter the City Name:")
Days=input("Enter the Number of Days:")
Style=input("Enter the Travel Style:")
Interest=input("What are Your Interests:")

# Initiate llm
llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash",google_api_key=userdata.get('GEMINI_API_KEY'))

Travel_template=PromptTemplate(template="""Create a Day-wise plan for the entered city {City} based on the the Travel Style {Style} and the Interests {Interest}.
Must generate the plan as:

Day 1
Day 2
… up to the number of {Days} entered by the user

2) Include time blocks for every day
Each day must contain:

Morning
Afternoon
Evening

3) Add one budget-saving tip per day
At the end of each day, Gemini must add:

✅ Budget Tip: (one practical money-saving suggestion)""",input_variables=["Days","City"])

# Travel_template => llm
# Chains
Travel_chain=Travel_template | llm | StrOutputParser()

# Invoke
response=Travel_chain.invoke({"City":City,"Days":Days,"Style":Style,"Interest":Interest})

# Print response
print(response)
```
